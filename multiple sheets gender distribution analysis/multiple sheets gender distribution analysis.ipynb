```python
import pandas as pd

records_file_name = 'Yiran-BurialRecords_CAA26Jun2019.xlsx'

df_records = pd.read_excel(io=records_file_name, sheet_name = None)
```

```python
for d in df_records:
    print(type(df_records[d]))
```

```python
_df = df_records['Jul 1929 - Feb 1930']
print(_df)
```

```python
_df['Sex'].value_counts()['M']
```

```python
results = []

for d in df_records:
    mf = df_records[d]['Sex'].value_counts()
    _sum = mf.sum()
    m = mf['M']
    f = mf['F']
    unknown = _sum - m - f
    results.append([d, _sum, m, f, unknown])
```

```python
print(results)
```

```python
import numpy as np
df_results = pd.DataFrame(np.array(results), columns=['Filenames', 'Line/names', 'Male', 'Female', 'Unknown'])
print(df_results)
```

```python
df_results.to_excel("output.xlsx")
```
